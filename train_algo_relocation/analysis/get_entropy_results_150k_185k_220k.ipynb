```python
import scipy.stats as sc

# here setup which model to evaluate
#1525336829 is the 150k expensive relocation model.
#1525445230 is the 185k expensive relocation model.
#1526820866 is the 220k expensive relocation model.

models_to_evaluate = [1525336829, 1525445230, 1526820866]

def load_failed_instances(model_nr):
    if model_nr == 1525336829:
        failed_DRL = failed_DRL_150k
        model_name = 'DRL_150k'
    elif model_nr == 1525445230:
        failed_DRL = failed_DRL_185k
        model_name = 'DRL_185k'
    elif model_nr == 1526820866:
        failed_DRL = failed_DRL_220k
        model_name = 'DRL_220k'
    else:
        print('wrong model nr entered, execution stopped')
    return failed_DRL, model_name


# Input a pandas series 
def ent(data):
    p_data= data.value_counts()/len(data) # calculates the probabilities
    entropy=sc.entropy(p_data)  # input probabilities to get the entropy 
    return entropy


records = []    
for model_nr in models_to_evaluate:
    failed_DRL, model_name = load_failed_instances(model_nr)
    action_data = pd.read_csv('data_'+str(model_nr)+'_relocation_arrival_actions.csv')
    
    # annotate solved instances
    action_data['solved'] = action_data.apply(lambda row: 0 if row.instance_id in failed_DRL else 1, axis = 1)
    
    #filter on solved instances
    action_data_solved = action_data.loc[action_data['solved'] == 1].copy()
    
    for type_ in action_data_solved.strtype.unique():
        actions = action_data_solved.loc[action_data_solved.strtype == type_].action.astype('int64').copy()
        ent_ = ent(actions)
        records.append({'type_': type_, 'algo': model_name, 'ent': ent_})
    
ent_data = pd.DataFrame.from_records(records)
ent_data['composition'] = ent_data.apply(lambda row: row.type_.replace('14','SLT4').replace('16', 'SLT6').replace('24','VIRM4').replace('26','VIRM6'), axis =1 )

result_table = ent_data.pivot(index='algo', columns='composition', values='ent')
result_table
```
